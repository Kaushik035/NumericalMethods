```python
import numpy as numpy
import matplotlib.pyplot as plt
import random
```

```python
def Vibonacci(n):
    """use a random number to decide if you want to add or subtract the next number, ploat it as function of iteration"""

    previous =1
    latest = 1

    series = []

    sum = 0
    while n > 0:
        randomNumber = random.random()
        # print("random number is ", randomNumber)
        if(randomNumber < 0.5):
            sum = latest + previous
        else:
            sum = latest - previous

        series.append(sum)
        previous = latest
        latest = sum
        n = n-1

    return series
        
```

```python
series = Vibonacci(10)
```

```python
x_values = list(range(1, len(series) + 1))  # [1, 2, ..., n]
y_values = series

plt.figure(figsize=(8, 4))
plt.plot(x_values, y_values, marker='o', linestyle='-', color='blue')
plt.title("Vibonacci Series")
plt.xlabel("Iteration")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```
